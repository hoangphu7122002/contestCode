```
def decentNumber(n):
  if n%5 == 0 and n%3 == 0:
      return int('5'*n)

  elif n%3 == 0:
      return int('5'*n)
  
  # If no match above, then n-digit-result is composed by
  # (3*b) digits of 5 and (5*a) digits of 3
  for a in range(0,n//5+1):
    if n - 5*a >= 0:
      if (n - 5*a)%3 == 0:
        b = int((n - 5*a)/3)
        return int('5'*(3*b) + '3'*(5*a))
    
  return -1

if __name__ == '__main__':
  t = int(input().strip())

  for t_itr in range(t):
    n = int(input().strip())
    print(decentNumber(n))
```
